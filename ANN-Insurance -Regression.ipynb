```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("ANN-Insurence.csv")
df
```

```python
df.info()
```

```python
#as we dont have null values so we can convert the categorical columns into numerical calumns
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
cat_cols="sex","smoker","region"
```

```python
for i in cat_cols:
    df[i]=le.fit_transform(df[i])
```

```python
df
```

```python
#so here we are converted the all columns as numerics
```

```python
#lets check the shape 
df.shape
```

```python
x=df.iloc[:,:-1]
x
```

```python
y=df.iloc[:,-1]
y
```

```python
#lets scale the data
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
x_scaler=scaler.fit_transform(x)
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_scaler,y,test_size=0.2)
```

```python
x_train.shape
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model=Sequential()
```

```python
model.add(Dense(12,input_shape=(6,),activation='relu'))
```

```python
model.add(Dense(8,activation='relu'))
```

```python
model.add(Dense(4,activation='relu'))
```

```python
model.add(Dense(1,activation='linear'))
```

```python
model.summary()
```

```python
model.compile(loss='mse',optimizer='adam',metrics=['mse','mae'])
```

```python
result=model.fit(x_train,y_train,batch_size=50,epochs=400)
```

```python
import matplotlib.pyplot as plt
```

```python
#evaluating the model
```

```python
model.evaluate(x_test,y_test)
```
